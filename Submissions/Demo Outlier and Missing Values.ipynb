```python
from sklearn.datasets import load_diabetes
```

```python
dataset = load_diabetes()
dataset
```

```python
dataset.data             # see data of the column 'data'
```

```python
dataset.target          # see data of the column 'target'
```

```python
dataset['feature_names']
```

```python
import pandas as pd
import numpy as np
```

```python
# convert into dataframe
df = pd.DataFrame(data=np.c_[dataset['data'],dataset['target']], columns=dataset['feature_names']+['target'])         
# np.c_ is to concatenate horizontally target and data values
df
```

```python
# check for missing data
df.isnull().any()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

for column in df:
    plt.figure()
    df.boxplot([column])
```
