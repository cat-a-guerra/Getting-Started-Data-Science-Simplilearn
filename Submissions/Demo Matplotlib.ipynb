```python
import numpy as np                 # for random numbers
import matplotlib.pyplot as plt    # plot numbers
from matplotlib import style       # sets grid style
%matplotlib inline                 
# display plot within Jupyter notebook
```

```python
# generate random numbers (10)
random_number = np.random.rand(10)
print(random_number)
```

```python
# select style of the plot
style.use('ggplot')

# plot the random number
plt.plot(random_number, 'g', label='line one', linewidth=2)        # set conditions

# x axis is number of random numbers (index)
plt.xlabel('Range')

# y axis is actual random number
plt.ylabel('Numbers')

# Title of the plot
plt.title('First plot')
plt.legend()       # plots graph based on set conditions
plt.show()         # displays graph
```

```python
import matplotlib.pyplot as plt    # plot numbers
%matplotlib inline         
```

```python
cause = 'Chronic Diseases', 'Unintentional Injuries', 'Alzheimers', 'Influenza and Pneumonia', 'Sepsis', 'Others'
percentile = [62, 5, 4, 2, 1, 26]
```

```python
# set plot properties

# set figure size
plt.figure(figsize=(10,10))

# slice the largest piece based on the percentile -> explode
explode = (0.05, 0, 0, 0, 0, 0)

# pie plot to pass the parameters
plt.pie(percentile, labels=cause, explode=explode, autopct='%1.1f%%', startangle=70) # autopct will embed the percentile value to each pie slice

# set axis equal to draw pie as circle
plt.axis('equal')

# set the title of the pie chart
plt.title('Ohio State: 2012 - Leading causes of death')

# show the plot
plt.show()
```
